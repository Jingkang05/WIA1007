```python
import numpy as np
# 1. Create a numpy array of size 2 x 2
arr_2x2 = np.array([[1, 2],[3, 4]])

print("--- 2x2 Array ---")
print(arr_2x2)
print("Shape:", arr_2x2.shape)
```

```python
# 2. Create a random numpy array of size 3 x 4
arr_rand_3x4 = np.random.rand(3, 4)

print(" 3x4 Random Array ")
print(arr_rand_3x4)
print("Shape:", arr_rand_3x4.shape)
```

```python
# 3. Create a numpy array of ones and a numpy array of zeroes
arr_ones = np.ones((3, 3))
arr_zeros = np.zeros((3, 3))

print(" 3x3 Array of Ones ")
print(arr_ones)

print(" 3x3 Array of Zeros ")
print(arr_zeros)
```

```python
# 4. Create a random numpy array of size 6x3 and transpose it
arr_6x3 = np.random.rand(6, 3)
arr_transposed = arr_6x3.T

print("\n--- Transposed 3x6 Array ---")
print(arr_transposed)
print("Transposed Shape:", arr_transposed.shape)
```

```python
# 5. Create two random numpy array of size n x p and p x m and then compute the dot product

# Let's define our dimensions:
n = 4
p = 2
m = 5

# Create the two random arrays
arr_nxp = np.random.rand(n, p)  # Shape (4, 2)
arr_pxm = np.random.rand(p, m)  # Shape (2, 5)

# Compute the dot product using np.dot()
dot_product = np.dot(arr_nxp, arr_pxm)

print("\n--- Dot Product ---")
print("Shape of first array:", arr_nxp.shape)
print("Shape of second array:", arr_pxm.shape)
print("\nResulting Dot Product:")
print(dot_product)
print("Result Shape:", dot_product.shape)
```
